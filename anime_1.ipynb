```python
#make a new column for episode count
def ep_no(txt):
    count=False
    data=""
    for i in txt:
        if i==")":
            count=False
            return data
            break     
        if count==True:
            data=data+i
        if i=="(":
            count=True

        
        
```

```python
df["Episodes"]=df['Title'].apply(ep_no)
```

```python
df["Episodes"]
```

```python
#make a new column for time stamp
df.loc[1]
```

```python

def aring_time(txt):
    count=False
    data=""
    for i in range(len(txt)):
        if txt[i]==")":
            for j in range(i+1,i+20):
                data=data+txt[j]

            return data
            
```

```python
df["aring_time"]=df["Title"].apply(aring_time)
```

```python
df["aring_time"]
```

```python
#which anime has the highest score
df[df["Score"]==df["Score"].max()]
```

```python
#give me top 5 highest scoring anime
df.sort_values(by="Score" ,ascending=0)
df.head()
```

```python
#which anime has the highest episode count
df['Episodes']=df["Episodes"].str.replace("eps","")
df["Episodes"]=df["Episodes"].astype(int)
```

```python
df[df["Episodes"]==df["Episodes"].max()]
```

```python
#animes with top 5 episode count
df=df.sort_values(by="Episodes",ascending=0)
```

```python
df.head()
```

```python
#which is the longest running anime
# Split into Start and End dates
df[['Start', 'End']] = df['aring_time'].str.split(' - ', expand=True)

# Convert to datetime
df['Start'] = pd.to_datetime(df['Start'], format='%b %Y')
df['End'] = pd.to_datetime(df['End'], format='%b %Y')

# Calculate duration in days
df['Duration_Days'] = (df['End'] - df['Start']).dt.days

# Convert to years (optional)
df['Duration_Years'] = np.round(df['Duration_Days'] / 365.25, 2)

# Find the longest-running anime
longest = df.loc[df['Duration_Days'].idxmax()]

print(longest[['Title', 'aring_time', 'Duration_Years']])
```
