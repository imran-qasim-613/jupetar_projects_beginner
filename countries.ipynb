```python
# Find the top 5 countries with the highest GDP per person.

# Hint:

# GDP per capita = gdp / population
df["gdp_per_cap"]=df['gdp']/df['population']
```

```python
df['gdp_per_cap']
```

```python
df=df.sort_values(by='gdp_per_cap',ascending=0)
df.head()
```

```python
# Create a new column called Population_Density.

# Hint:

# Population Density = population / land_area
df['population_density']=df['population']/df['land_area']
```

```python
df['population_density']
```

```python
# Find the country with the highest population density
print(df.loc[df['population_density'].idxmax()]['country'])

```

```python
# Find the continent with the highest average GDP.
df.groupby('continent')['gdp'].mean()
```

```python
# Find the top 10 countries with both:
# High democracy score
# High life expectancy

df.sort_values(by=['democracy_score', 'life_expectancy'],ascending=False).head(10)
```

```python
# Find the correlation between:
# GDP and Life Expectancy
# GDP and Internet Usage
df['gdp'].corr(df['life_expectancy'])
```
